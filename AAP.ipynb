# Fitting the model
- Finally fitting the model after splitting the data set

```python
y=finalData["loan_status"].fillna(finalData["loan_status"].mean())
x=finalData.drop(['id', 'funded_amnt', 'funded_amnt_inv', 'dti', 'inq_last_6mths', 'mths_since_last_record', 
        'revol_bal', 'revol_util', 'out_prncp',
       'out_prncp_inv', 'total_pymnt_inv', 'total_rec_prncp',
       'total_rec_int', 'recoveries',
       'collection_recovery_fee', 'last_pymnt_amnt',
       'collections_12_mths_ex_med', 'pub_rec_bankruptcies','home_ownership',
        'emp_title',
       'verification_status', 'purpose', 'title'], axis =1).fillna(finalData.mean())
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
score = logmodel.score(x_test, y_test)
print(score)
```
